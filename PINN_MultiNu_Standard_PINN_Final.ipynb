```python
# ============================================================
# Final Standard PINN for Sharp-Gradient Second-Order ODE
#
# Problem:
#   u''(x) - Pe*u'(x) = 0,   0 <= x <= 1
#   u(0) = 1, u(1) = 0
#
# Exact solution:
#   u(x) = (1 - exp(Pe*(x - 1))) / (1 - exp(-Pe))
#
# Cases:
#   nu = 1, 0.1, 0.01, 0.001
#   Pe = 1, 10, 100, 1000
#
# Method:
#   Standard soft-boundary PINN
#   Adam + L-BFGS
# ============================================================

import os
import time
import random
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn


# ============================================================
# 1. Reproducibility and device
# ============================================================

SEED = 1234

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Use float32 for Colab GPU speed.
# This does not change the method.
DTYPE = torch.float32
torch.set_default_dtype(DTYPE)

print("CUDA available:", torch.cuda.is_available())
print("Device:", DEVICE)

if torch.cuda.is_available():
    print("GPU name:", torch.cuda.get_device_name(0))

print("Torch dtype:", torch.get_default_dtype())


# ============================================================
# 2. Problem cases
# ============================================================

CASES = [
    {"nu": 1.0, "Pe": 1.0},
    {"nu": 0.1, "Pe": 10.0},
    {"nu": 0.01, "Pe": 100.0},
    {"nu": 0.001, "Pe": 1000.0},
]

U_LEFT = 1.0
U_RIGHT = 0.0


# ============================================================
# 3. Training settings
# ============================================================

N_COLLOCATION = 10000

ADAM_EPOCHS = 15000
ADAM_LR = 1.0e-3

LBFGS_MAX_ITER = 500
LBFGS_LR = 1.0

BOUNDARY_WEIGHT = 100.0

PRINT_EVERY = 500

MAKE_PLOTS = True
SAVE_RESULTS = True

OUTPUT_DIR = "pinn_standard_results"


# ============================================================
# 4. Exact solution
# ============================================================

def exact_solution_np(x, Pe):
    x = np.asarray(x, dtype=np.float64)

    if abs(Pe) < 1.0e-14:
        return 1.0 - x

    return (1.0 - np.exp(Pe * (x - 1.0))) / (1.0 - np.exp(-Pe))


# ============================================================
# 5. Neural network model
# ============================================================

class PINN(nn.Module):
    def __init__(self, width=64, depth=4):
        super().__init__()

        layers = []

        layers.append(nn.Linear(1, width))
        layers.append(nn.Tanh())

        for _ in range(depth - 1):
            layers.append(nn.Linear(width, width))
            layers.append(nn.Tanh())

        layers.append(nn.Linear(width, 1))

        self.net = nn.Sequential(*layers)

        self.initialize_weights()

    def initialize_weights(self):
        for layer in self.net:
            if isinstance(layer, nn.Linear):
                nn.init.xavier_normal_(layer.weight)
                nn.init.zeros_(layer.bias)

    def forward(self, x):
        return self.net(x)


# ============================================================
# 6. Grids
# ============================================================

def make_collocation_points(n_points):
    x = torch.linspace(
        0.0,
        1.0,
        n_points,
        device=DEVICE,
        dtype=DTYPE
    ).view(-1, 1)

    x.requires_grad_(True)

    return x


def make_dense_grid(Pe):
    uniform_part = np.linspace(0.0, 1.0, 2001)

    layer_distance = np.geomspace(1.0e-14, 1.0, 5000)
    layer_part = 1.0 - layer_distance

    x_dense = np.unique(
        np.concatenate(
            [
                uniform_part,
                layer_part,
                np.array([0.0, 1.0])
            ]
        )
    )

    x_dense = x_dense[(x_dense >= 0.0) & (x_dense <= 1.0)]
    x_dense.sort()

    return x_dense


def make_table_grid(Pe):
    standard = np.linspace(0.0, 1.0, 11)

    layer_points = np.array(
        [
            1.0 - 5.0 / Pe,
            1.0 - 2.0 / Pe,
            1.0 - 1.0 / Pe,
            1.0 - 0.5 / Pe,
            1.0 - 0.2 / Pe,
            1.0 - 0.1 / Pe,
            1.0 - 0.05 / Pe,
            1.0 - 0.01 / Pe,
            1.0,
        ],
        dtype=np.float64
    )

    layer_points = layer_points[(layer_points >= 0.0) & (layer_points <= 1.0)]

    x_table = np.unique(np.concatenate([standard, layer_points]))
    x_table.sort()

    return x_table


# ============================================================
# 7. PINN loss
# ============================================================

def compute_pinn_losses(model, x_col, Pe):
    u = model(x_col)

    u_x = torch.autograd.grad(
        outputs=u,
        inputs=x_col,
        grad_outputs=torch.ones_like(u),
        create_graph=True
    )[0]

    u_xx = torch.autograd.grad(
        outputs=u_x,
        inputs=x_col,
        grad_outputs=torch.ones_like(u_x),
        create_graph=True
    )[0]

    residual = u_xx - Pe * u_x
    residual_loss = torch.mean(residual ** 2)

    x0 = torch.tensor([[0.0]], device=DEVICE, dtype=DTYPE)
    x1 = torch.tensor([[1.0]], device=DEVICE, dtype=DTYPE)

    u0 = model(x0)
    u1 = model(x1)

    boundary_loss = (u0 - U_LEFT) ** 2 + (u1 - U_RIGHT) ** 2
    boundary_loss = torch.mean(boundary_loss)

    total_loss = residual_loss + BOUNDARY_WEIGHT * boundary_loss

    return total_loss, residual_loss, boundary_loss


# ============================================================
# 8. Train one case
# ============================================================

def train_pinn_for_case(nu, Pe):
    print("\n============================================================")
    print(f"Training Standard PINN for Pe={Pe:g}, nu={nu:g}")
    print("============================================================")

    start_time = time.time()

    model = PINN(width=64, depth=4).to(DEVICE)
    x_col = make_collocation_points(N_COLLOCATION)

    optimizer_adam = torch.optim.Adam(model.parameters(), lr=ADAM_LR)

    adam_history = []

    for epoch in range(ADAM_EPOCHS + 1):
        optimizer_adam.zero_grad()

        total_loss, residual_loss, boundary_loss = compute_pinn_losses(
            model,
            x_col,
            Pe
        )

        total_loss.backward()
        optimizer_adam.step()

        if epoch % PRINT_EVERY == 0:
            adam_history.append(
                [
                    epoch,
                    total_loss.item(),
                    residual_loss.item(),
                    boundary_loss.item()
                ]
            )

            print(
                f"Adam Epoch {epoch:6d} | "
                f"Total = {total_loss.item():.6e} | "
                f"Residual = {residual_loss.item():.6e} | "
                f"BC = {boundary_loss.item():.6e}"
            )

    optimizer_lbfgs = torch.optim.LBFGS(
        model.parameters(),
        lr=LBFGS_LR,
        max_iter=LBFGS_MAX_ITER,
        max_eval=LBFGS_MAX_ITER,
        tolerance_grad=1.0e-12,
        tolerance_change=1.0e-12,
        history_size=100,
        line_search_fn="strong_wolfe"
    )

    lbfgs_counter = {"iter": 0}

    def closure():
        optimizer_lbfgs.zero_grad()

        total_loss, residual_loss, boundary_loss = compute_pinn_losses(
            model,
            x_col,
            Pe
        )

        total_loss.backward()

        lbfgs_counter["iter"] += 1

        if lbfgs_counter["iter"] % 50 == 0:
            print(
                f"L-BFGS Iter {lbfgs_counter['iter']:5d} | "
                f"Total = {total_loss.item():.6e} | "
                f"Residual = {residual_loss.item():.6e} | "
                f"BC = {boundary_loss.item():.6e}"
            )

        return total_loss

    print("\nStarting L-BFGS refinement...")
    optimizer_lbfgs.step(closure)

    final_total_loss, final_residual_loss, final_boundary_loss = compute_pinn_losses(
        model,
        x_col,
        Pe
    )

    training_time = time.time() - start_time

    print("\nFinal training losses")
    print(f"Total loss        = {final_total_loss.item():.6e}")
    print(f"Residual loss     = {final_residual_loss.item():.6e}")
    print(f"Boundary loss     = {final_boundary_loss.item():.6e}")
    print(f"Training time [s] = {training_time:.2f}")

    train_info = {
        "adam_history": np.array(adam_history, dtype=np.float64),
        "final_total_loss": final_total_loss.item(),
        "final_residual_loss": final_residual_loss.item(),
        "final_boundary_loss": final_boundary_loss.item(),
        "training_time": training_time
    }

    return model, train_info


# ============================================================
# 9. Evaluate trained model
# ============================================================

def evaluate_model(model, Pe):
    model.eval()

    x_dense = make_dense_grid(Pe)

    x_torch = torch.tensor(
        x_dense.reshape(-1, 1),
        device=DEVICE,
        dtype=DTYPE
    )

    with torch.no_grad():
        u_pred = model(x_torch).detach().cpu().numpy().reshape(-1)

    u_exact = exact_solution_np(x_dense, Pe)
    abs_error = np.abs(u_pred - u_exact)

    max_error = np.max(abs_error)
    mean_error = np.mean(abs_error)
    rms_error = np.sqrt(np.mean(abs_error ** 2))
    l2_error = np.sqrt(np.trapezoid(abs_error ** 2, x_dense))

    x0 = torch.tensor([[0.0]], device=DEVICE, dtype=DTYPE)
    x1 = torch.tensor([[1.0]], device=DEVICE, dtype=DTYPE)

    with torch.no_grad():
        u0 = model(x0).item()
        u1 = model(x1).item()

    left_boundary_error = abs(u0 - U_LEFT)
    right_boundary_error = abs(u1 - U_RIGHT)

    return {
        "x_dense": x_dense,
        "u_pred": u_pred,
        "u_exact": u_exact,
        "abs_error": abs_error,
        "max_error": max_error,
        "mean_error": mean_error,
        "rms_error": rms_error,
        "l2_error": l2_error,
        "left_boundary_error": left_boundary_error,
        "right_boundary_error": right_boundary_error
    }


# ============================================================
# 10. Prediction table
# ============================================================

def make_prediction_table(model, Pe):
    model.eval()

    x_table = make_table_grid(Pe)

    x_torch = torch.tensor(
        x_table.reshape(-1, 1),
        device=DEVICE,
        dtype=DTYPE
    )

    with torch.no_grad():
        u_pred = model(x_torch).detach().cpu().numpy().reshape(-1)

    u_exact = exact_solution_np(x_table, Pe)
    abs_error = np.abs(u_pred - u_exact)

    table = np.column_stack(
        [
            x_table,
            u_pred,
            u_exact,
            abs_error
        ]
    )

    return table


# ============================================================
# 11. Main program
# ============================================================

def main():
    if SAVE_RESULTS:
        os.makedirs(OUTPUT_DIR, exist_ok=True)

    summary_rows = []

    for case in CASES:
        nu = float(case["nu"])
        Pe = float(case["Pe"])

        model, train_info = train_pinn_for_case(nu=nu, Pe=Pe)

        eval_info = evaluate_model(model, Pe)
        table = make_prediction_table(model, Pe)

        print("\nTable: PINN vs Exact")
        print("x              PINN solution         Exact solution        Absolute error")

        for row in table:
            x, numerical, exact, error = row

            print(
                f"{x: .10f}     "
                f"{numerical: .12e}     "
                f"{exact: .12e}     "
                f"{error: .4e}"
            )

        print("\nDense-grid error metrics")
        print(f"Maximum absolute error = {eval_info['max_error']:.6e}")
        print(f"Mean absolute error    = {eval_info['mean_error']:.6e}")
        print(f"RMS error              = {eval_info['rms_error']:.6e}")
        print(f"L2-type error          = {eval_info['l2_error']:.6e}")
        print(f"Left boundary error    = {eval_info['left_boundary_error']:.6e}")
        print(f"Right boundary error   = {eval_info['right_boundary_error']:.6e}")

        tag = f"pe{int(Pe)}"

        if SAVE_RESULTS:
            dense_rows = np.column_stack(
                [
                    eval_info["x_dense"],
                    eval_info["u_pred"],
                    eval_info["u_exact"],
                    eval_info["abs_error"]
                ]
            )

            np.savetxt(
                os.path.join(OUTPUT_DIR, f"pinn_{tag}_dense.csv"),
                dense_rows,
                delimiter=",",
                header="x,u_pinn,u_exact,absolute_error",
                comments=""
            )

            np.savetxt(
                os.path.join(OUTPUT_DIR, f"pinn_{tag}_table.csv"),
                table,
                delimiter=",",
                header="x,u_pinn,u_exact,absolute_error",
                comments=""
            )

            np.savetxt(
                os.path.join(OUTPUT_DIR, f"pinn_{tag}_adam_history.csv"),
                train_info["adam_history"],
                delimiter=",",
                header="epoch,total_loss,residual_loss,boundary_loss",
                comments=""
            )

            torch.save(
                model.state_dict(),
                os.path.join(OUTPUT_DIR, f"pinn_{tag}_model.pt")
            )

        if MAKE_PLOTS:
            plt.figure(figsize=(7, 4.5))
            plt.plot(eval_info["x_dense"], eval_info["u_pred"], label="PINN")
            plt.plot(eval_info["x_dense"], eval_info["u_exact"], "--", label="Exact")
            plt.xlabel("x")
            plt.ylabel("u(x)")
            plt.title(f"Standard PINN vs Exact, Pe={Pe:g}")
            plt.legend()
            plt.tight_layout()

            if SAVE_RESULTS:
                plt.savefig(
                    os.path.join(OUTPUT_DIR, f"pinn_{tag}_solution.png"),
                    dpi=300,
                    bbox_inches="tight"
                )

            plt.show()

            plt.figure(figsize=(7, 4.5))
            plt.semilogy(
                eval_info["x_dense"],
                eval_info["abs_error"] + 1.0e-18
            )
            plt.xlabel("x")
            plt.ylabel("absolute error")
            plt.title(f"Absolute Error, Standard PINN, Pe={Pe:g}")
            plt.tight_layout()

            if SAVE_RESULTS:
                plt.savefig(
                    os.path.join(OUTPUT_DIR, f"pinn_{tag}_error.png"),
                    dpi=300,
                    bbox_inches="tight"
                )

            plt.show()

        summary_rows.append(
            [
                Pe,
                nu,
                N_COLLOCATION,
                64,
                4,
                ADAM_EPOCHS,
                LBFGS_MAX_ITER,
                BOUNDARY_WEIGHT,
                train_info["final_total_loss"],
                train_info["final_residual_loss"],
                train_info["final_boundary_loss"],
                train_info["training_time"],
                eval_info["max_error"],
                eval_info["mean_error"],
                eval_info["rms_error"],
                eval_info["l2_error"],
                eval_info["left_boundary_error"],
                eval_info["right_boundary_error"]
            ]
        )

    summary_rows = np.array(summary_rows, dtype=np.float64)

    print("\n============================================================")
    print("Final PINN Summary")
    print("============================================================")
    print(
        "Pe        nu        Ncol      width depth  "
        "total loss     residual loss  BC loss       "
        "max err       mean err      RMS err       L2 err"
    )

    for row in summary_rows:
        Pe = row[0]
        nu = row[1]
        ncol = row[2]
        width = row[3]
        depth = row[4]
        total_loss = row[8]
        residual_loss = row[9]
        boundary_loss = row[10]
        max_err = row[12]
        mean_err = row[13]
        rms_err = row[14]
        l2_err = row[15]

        print(
            f"{Pe:7.1f}  "
            f"{nu:8.4g}  "
            f"{int(ncol):7d}  "
            f"{int(width):5d} "
            f"{int(depth):5d}  "
            f"{total_loss: .3e}  "
            f"{residual_loss: .3e}  "
            f"{boundary_loss: .3e}  "
            f"{max_err: .3e}  "
            f"{mean_err: .3e}  "
            f"{rms_err: .3e}  "
            f"{l2_err: .3e}"
        )

    if SAVE_RESULTS:
        summary_path = os.path.join(
            OUTPUT_DIR,
            "pinn_standard_multiple_pe_summary.csv"
        )

        np.savetxt(
            summary_path,
            summary_rows,
            delimiter=",",
            header=(
                "Pe,nu,N_collocation,width,depth,adam_epochs,lbfgs_max_iter,"
                "boundary_weight,total_loss,residual_loss,boundary_loss,"
                "training_time,max_absolute_error,mean_absolute_error,"
                "rms_error,l2_error,left_boundary_error,right_boundary_error"
            ),
            comments=""
        )

        print("\nSaved files inside folder:")
        print(f"  {OUTPUT_DIR}/")
        print("\nMain summary file:")
        print(f"  {summary_path}")


main()
```
